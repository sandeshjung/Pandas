```python
def calculate_profit(unit_cost, unit_price,units_sold):
    if isinstance(unit_cost,(int,float)) and isinstance(unit_price,(int,float)) and isinstance(units_sold,(int,float)):
        return units_sold*(unit_price-unit_cost)
    else:
        return None
```

```python
total_profit = sales.apply(lambda x: calculate_profit(x['Unit Cost'],x['Unit Price'],x['Units Sold']),axis=1)
```

```python
non_null_indices = ~sales[['Unit Cost','Unit Price','Total Profit']].isna().any(axis=1)
```

```python
pd.concat([total_profit[non_null_indices], sales['Total Profit'][non_null_indices]], axis=1).head()
```

```python
#Pipe method to chain sequence of processing steps
```

```python
def item_to_lower(df):
    df=df.copy()
    df_['Item Type'] = df_['Item Type'].str.lower()
    return df_
def get_total_items_sold(df):
    df_=df.copy()
    df_=df_.groupby('Item Type',as_index=False).agg({'Units Sold':sum})
    return df_
```

```python
sales.pipe(item_to_lower).pipe(get_total_items_sold)
```
